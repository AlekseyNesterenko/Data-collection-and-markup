```python
import requests
import json
import pandas as pd 
import datetime as dt 
```

```python
url = 'http://api.binance.us/api/v3/klines'
symbol = 'BTCBUSD'
interval = '1h'
start = str(int(dt.datetime(2021,1,1).timestamp()*1000))
end = str(int(dt.datetime(2021,12,1).timestamp()*1000))
```

```python
params = {
    'symbol': symbol,
    'interval': interval,
    'startTime': start,
    'endTime': end
}
```

```python
responce = requests.get(url, params= params).text
```

```python
responce
```

```python
json_res = json.loads(responce)
```

```python
df = pd.DataFrame(json_res)
```
